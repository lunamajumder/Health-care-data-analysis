```python
# @title Default title text

# Import pandas
import pandas as pd

# Load the CSV file from Google Drive
df = pd.read_csv('/content/drive/MyDrive/total.csv')
```

```python
# @title Unplanned Readmission in different Measure

df.groupby('Measure').size().sort_values(ascending=True).plot(kind='barh', color='skyblue')

# Clean up plot appearance
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Unplanned Readmissions by Measure')
plt.xlabel('Number of Readmissions')
plt.ylabel('Measure')
plt.tight_layout()
plt.show()
```

```python
# @title Readmission to Same Hospital or not

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Readmission_to_Same_Hospital_or_not').size().plot(kind='barh')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('Readmission to Same Hospital or not')


```

```python
# @title Pneumonia patients had the longest average length of stay during the initial admission.

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Measure'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Admission.LOS', y='Measure', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.title('Pneumonia patients had the longest average length of stay during the initial admission.')



```

```python
# @title Among unplanned readmissions, Stroke patients had the longest average length of stay, followed by those with COPD, Pneumonia & HF.
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Measure'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Readmission.LOS', y='Measure', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.title('Among unplanned readmissions, Stroke patients had the longest average length of stay, followed by those with COPD, Pneumonia & Heart Failure')
```

```python
# @title A strong correlation was observed between the length of stay during the initial admission and that of the unplanned readmission.

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Admission.LOS', y='Readmission.LOS', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('A strong correlation was observed between the length of stay during the initial admission and that of the unplanned readmission.')
```
